```python
streams = pd.read_csv('Streams_Index.csv')
streams
```

# 3. The Graph
The data shown takes place between the days of June 2nd, 2015 and April 20th, 2021 (2,150 days).
The 1.0 represents the day Quadeca got the most streams.

```python
streams.plot(y='streams_index', x='date')
plt.rcParams["figure.figsize"] = (20,12)
```

# 4. Analyzing the Graph
* Quadeca's most streamed day was on March 31st, 2021. The reason that he had a spike in streams that day is because his album, titled "From Me To You", was released the night before.
* Quadeca's second largest spike happened on March 14th, 2019. That was six (6) days after his "breakthrough" album, titled "Voice Memos", was released.
* In late-2018, Quadeca started getting more streams due to exposure from a YouTuber and rapper named KSI.
* Quadeca's streams in 2020 were very up and down for two main reasons: 1. He had released three songs (causing the rise in streams, because there is new music), 2. He was took a roughly five-month hiatus from social media and YouTube (causing the decline in streams, because there was no new content).
